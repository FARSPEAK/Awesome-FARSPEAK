```python
import time
import requests
from google.colab import files
```

```python
# Allow user to upload a file
uploaded = files.upload()

# Get the file name of the uploaded file
file_name = list(uploaded.keys())[0]

### TODO: 4 variables from the farspeak dashboard, including the token for the app/environment
farspeak_appname = ''
farspeak_env = ''
farspeak_entity_name = ''
farspeak_token = ''  # Replace with actual token

url = f'https://api.farspeak.ai/docs/{farspeak_appname}/{farspeak_env}/{farspeak_entity_name}'


# Define the form-data payload
form_data = {
    'template': '{ "full_name": "First and last name of the paper author", "insight": "The key insight of the paper" }',
    'instructions': 'This is a research paper.'
}

# Read the uploaded file content
file_content = uploaded[file_name]

# Define the files payload
files_payload = {
    'docs': (file_name, file_content, 'application/pdf')
}

# Define the headers for the request
headers = {
    'token': farspeak_token
}

# Make the POST request
response = requests.post(url, headers=headers, data=form_data, files=files_payload)

# Print the response from the server
print('Status code:', response.status_code)
print('Response text:', response.text)

```

```python
response_data = response.json()
job_id = response_data['ids'][0]

print("Job ID: " + job_id)

# Define the status URL
status_url = f'https://api.farspeak.ai/apps/{farspeak_appname}/{farspeak_env}/{farspeak_entity_name}'

# Function to check job status
def check_job_status(url, headers):
    response = requests.get(url, headers=headers)
    return response.status_code, response.text

# Poll the status until the job is completed
while True:
    status_code, status_text = check_job_status(status_url, headers)
    print(f'Status code: {status_code}')
    print(f'Status text: {status_text}')

    # Check if job is completed
    if status_code == 200:
        break
    # Wait before checking again
    time.sleep(5)
```

```python
# Query the completed job
query_url = f'https://api.farspeak.ai/knowledgebase/{farspeak_appname}/{farspeak_env}/{farspeak_entity_name}'
query_payload = {
    'query': "What's the main insight of this paper?"
}

# Send the query request
query_response = requests.post(query_url, headers=headers, json=query_payload)

# Print the query response
print('Query status code:', query_response.status_code)
print('Query response text:', query_response.text)
```
